## SpmImages.jl Demo

```julia
using SpmImages
```

### Load SPM image

```julia
ima = load_image("Image_445.sxm");
```

## Plot a specific image channel directly from the image

```julia
plot_channel(ima, "frequency shift", pixel_units=false)
```

### Get a channel as 2D data and directly plot it

```julia
c = get_channel(ima, "current");
plot_data(c.data, color=:lapaz, title="current", legend=:none)
```

### Plot using the Images library

```julia
using Images
d = get_channel(ima, "current bwd", origin="upper").data
d = d .- minimum(d)
d = d ./ maximum(d)
Gray.(d)
```

### Plot using the Imageview library

```julia
using ImageView
d = get_channel(ima, "frequency shift fwd", origin="upper").data
imshow(d, axis=(2,1));
```
